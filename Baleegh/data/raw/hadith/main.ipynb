```python
import json
import pandas as pd
```

```python
with open("./abuaminaelias_scraper/blockquotes.json") as file:
    data = json.load(file)
```

```python
print(data[5]['blockquote'])
```

```python
from bs4 import BeautifulSoup
```

```python
df = {"Arabic": [],
      "English": []
      }
```

```python
for i in range(len(data)):
    soup = BeautifulSoup(data[i]['blockquote'], 'html.parser')
    paragraphs = soup.find_all('p')

    if len(paragraphs) >= 2:
        first_paragraph = paragraphs[0].text
        second_paragraph = paragraphs[1].text
        
        df["Arabic"].append(first_paragraph)
        df["English"].append(second_paragraph)
        
        print(paragraphs[2:])
    else:
        print("There are fewer than two paragraphs in the blockquote.")
```

```python
df = pd.DataFrame(df)
```

```python
df.sample().values
```

```python
df = df[~df['Arabic'].str.contains(r'[a-zA-Z]', na=False)]
df = df[~df['English'].str.contains(r'[a-zA-Z]', na=False) == False]
```

```python
df.sample().values
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
df.to_csv("abuamina_hadith.csv")
```
